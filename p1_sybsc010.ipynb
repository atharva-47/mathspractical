```python
# Q1 : To  plot the function f(x) = log10(x) in the interval [0,10].
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,10)
y=np.log10(x)
plt.plot(x,y)
plt.show()
```

```python
#Q2 . :To plot the function f(x)=cos(x) in the interval [0.2pi]
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(0,2*(np.pi),0.1)
y=np.cos(x)
plt.plot(x,y)
plt.show()
```

```python
# Q4. To plot the function f(x)=x^2 in the interval [-2,2]
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Q6. To plot the function f(x)=sin(x)-e^x+3x^2-log10(x) in the interval [0,2pi]
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(0,2*(np.pi),0.1)
y=np.sin(x)-np.exp(x)+3*x**2+np.log10(x)
plt.plot(x,y)
plt.show()
```
