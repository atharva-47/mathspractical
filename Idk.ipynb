```python
from mpl_toolkits import *
from numpy import *
from pylab import *
a=[
    (70, -25, 15),
    (50, 72, -45),
    (58, -82, 65),
    (20, 72, -45)    
]


b=['D','*','.','*']
c=['black','red','green','green']
ax=axes(projection='3d')
for i in range(len(a)):
    x,y,z=a[i]
    color=c[i]
    marker=b[i]
    ax.scatter(x,y,z,c=color,marker=marker)

plt.show()
```

```python
#Write a python program in 3D to rotate the point (1, 0, 0) through XZ plane in clockwise direction
#(rotation through Y axis by an angle of 90).

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import *


p = np.array([1, 0, 0])


R = np.array([[1, 0, 0], [0, 1, 0], [0, 0, -1]])

p_rotated = R.dot(p)

ax=axes(projection='3d')
ax.scatter(*p_rotated, color='blue')
ax.plot([0, 0], [-1, 1], [0, 0], alpha=0.2)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')


plt.show()
```

```python

import matplotlib.pyplot as plt
vertices = [[5, 4], [7, 4], [6, 6]]
x, y = zip(*vertices)
plt.plot(x, y, 'g-', *zip(vertices[0], vertices[1]), 'g-', *zip(vertices[1], vertices[2]), 'g-', *zip(vertices[2], vertices[0]), 'g-')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


x = np.linspace(10, 20, 100)
y = np.sin(2 * x)
z = x
x = np.cos(2 * x)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, color='red')
ax.set_title('3D Line Graph')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.show()
```
